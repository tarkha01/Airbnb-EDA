1.Identify Key Features and Variables
Based on the analysis, the following variables are identified as the primary drivers of user behavior:

 country_destination: The target variable (The outcome we want to predict).

 age: Significant demographic differentiator, with most users in the 25-40 range.

 first_device_type: Reflects user tech-preferences (Apple vs. Android/Windows).

 signup_method & affiliate_channel: Key for understanding how marketing impacts acquisition.

 date_account_created: Essential for analyzing growth trends over time.

2.Main Insights
  High Churn/Window Shopping: A vast majority of users create accounts but never book (the NDF category).

  Platform Loyalty: Most users join "Directly" rather than through paid ads, indicating strong brand equity.

  Technical Profile: Mac and iPhone users represent the largest and most active segment of the user base.

  Domestic Focus: For those who do book, the US is the primary destination by a significant margin.

3.The Narrative: "The Tech-Savvy Window Shopper"
  The data reveals a platform that has mastered user acquisition but is still perfecting user conversion. We see a young, affluent, and tech-savvy audience (primarily Apple users) flocking to the platform in an exponential growth pattern between 2012 and 2014. However, a large percentage of these users are signing up and browse, but the journey often ends at account creation without a booking. We want to convert these high-intent browsers into travelers.

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

df_train['date_account_created'] = pd.to_datetime(df_train['date_account_created'])

# PLOT 1: The Conversion Problem (Target Destination Distribution)

dest_counts = df_train['country_destination'].value_counts().reset_index()
dest_counts.columns = ['Destination', 'Count']

fig1 = px.bar(dest_counts, x='Destination', y='Count',
              title="1. The Conversion Gap: Bookings vs. Non-Bookers (NDF)",
              text='Count', color='Destination',
              color_discrete_sequence=px.colors.qualitative.Pastel)
fig1.update_traces(textposition='outside')
fig1.update_layout(xaxis_title="Country Destination", yaxis_title="Number of Users",
                   showlegend=False, template="plotly_white")
fig1.show()
```

```python
# PLOT 2: Exponential User Growth Over Time

df_train['YearMonth'] = df_train['date_account_created'].dt.to_period('M').astype(str)
growth_df = df_train.groupby('YearMonth').size().reset_index(name='New Accounts')

fig2 = px.line(growth_df, x='YearMonth', y='New Accounts', markers=True,
               title="2. Platform Adoption: Exponential Account Growth (2010 - 2014)",
               labels={'YearMonth': 'Account Creation Date', 'New Accounts': 'Number of Signups'})
fig2.update_traces(line_color='#FF5A5F', line_width=3) # Airbnb brand color
fig2.update_layout(template="plotly_white")
fig2.show()
```

```python
# PLOT 3: The Millennial Core (Age Distribution Filtered 15-100 like an EDA)

age_filtered_df = df_train[(df_train['age'] >= 15) & (df_train['age'] <= 100)]

fig3 = px.histogram(age_filtered_df, x="age", nbins=40,
                    title="3. Demographics: The Millennial Core (Ages 15-100)",
                    labels={'age': 'User Age'}, color_discrete_sequence=['#00A699'])
fig3.update_layout(yaxis_title="Count of Users", template="plotly_white", bargap=0.1)
fig3.show()
```

```python
# PLOT 4: Tech-Savvy Audience (First Device Type)

device_counts = df_train['first_device_type'].value_counts().reset_index()
device_counts.columns = ['Device Type', 'Count']
device_counts = device_counts[device_counts['Count'] > 2000]

fig4 = px.pie(device_counts, names='Device Type', values='Count', hole=0.4,
              title="4. User Ecosystem: Dominance of Mac & Apple Devices",
              color_discrete_sequence=px.colors.qualitative.Set3)
fig4.update_traces(textposition='inside', textinfo='percent+label')
fig4.update_layout(template="plotly_white")
fig4.show()
```
